1. What is a random variable in probability theory ?

- A random variable is a numerical outcome of a random phenomenon. It assigns a real number to each outcome in a sample space.

2. What are the types of random variables ?

- Discrete Random Variable: Takes countable values (e.g., 0, 1, 2…).

- Continuous Random Variable: Takes values from a continuous range (e.g., any real number between 0 and 1).

3.  What is the difference between discrete and continuous distributions ?

- Discrete distribution describes probabilities for discrete random variables.

- Continuous distribution describes probabilities using a density function; probabilities are over intervals, not individual points.

4.  What are probability distribution functions (PDF) ?

- A probability distribution function gives the probability that a random variable takes a particular value (for discrete variables) or the relative likelihood over a range (for continuous variables).

- For discrete variables: It’s often called the probability mass function (PMF).

- For continuous variables: It's the probability density function (PDF).

6. What is a discrete uniform distribution ?

- A discrete uniform distribution assigns equal probability to all outcomes in a finite set. Example: Rolling a fair die where each number (1–6) has probability 1/6.

13.    What is the Central Limit Theorem (CLT), and why is it critical in statistics ?

- CLT states that the sampling distribution of the sample mean approaches a normal distribution as the sample size increases, regardless of the original distribution.
It enables statistical inference using the normal model.

- Z-statistics is used to determine how many standard deviations an observed value is from the mean. It's used in:

17.  What are point estimates and interval estimates in statistics ?

- Point estimate: Single value estimate of a parameter (e.g., sample mean).

- Interval estimate: Range of values (like a confidence interval) likely to contain the population parameter.

18.  What is the significance of confidence intervals in statistical analysis ?

- A confidence interval gives a range of plausible values for a population parameter with a specified level of confidence (e.g., 95%). It quantifies uncertainty in estimation.
